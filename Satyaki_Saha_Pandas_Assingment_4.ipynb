```
# Q2.- Make a function for batting profile of any batsman:-                                              [Medium+]

def details(name):
    mask1=df1['batsman']==name
    mask2=df1['bowler']==name
    mask3=df1['non_striker']==name
    mask4=df1['fielder']==name
    m=df1[mask1 | mask2 | mask3 | mask4]
    q=m[m['batsman']==name]
    s=df.groupby('player_of_match')
    l=s.get_group(name)
    mask4=m['batsman_runs']==4
    mask5=m['batsman_runs']==6
    m1=m[(mask4 | mask5)& mask1]
    inning=q.drop_duplicates(subset=['match_id']).count()['inning']
    print(name)
    print("Innings:",inning)
    print("Total runs:",q['batsman_runs'].sum())
    print("Man of the match:",l['id'].count())
    print("Boundaries:",m1['batsman_runs'].count())
```

```
# Q3.- Write a function to return top-4 batsman at a venue. Take venue as input.
```

```
new= df.merge(df1,left_on='id',right_on='match_id')
def ven(v):
    a=new[new['venue']==v]
    return a.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(4)
```

```
ven('Rajiv Gandhi International Stadium, Uppal')
```
